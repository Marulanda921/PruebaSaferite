```python
import requests
import pandas as pd
from tabulate import tabulate

basePokemon= "https://pokeapi.co/api/v2/pokemon/"
baseHabitat = 'https://pokeapi.co/api/v2/pokemon-species/'
baseMovimiento = 'https://pokeapi.co/api/v2/move/'
idsPokemons = [25,52,124]
listaMovimientos = []


def obtenerInfoPokemon(idPokemon):
    url = f'{basePokemon}{idPokemon}/'
    respuesta = requests.get(url)

    if respuesta.status_code != 200:
        return {'Error': 'no hay pokemones'}
    datos = respuesta.json()
    nombre = datos['name']
    
    movimientos = []
    for infoMovimiento in datos['moves']:
        movimientos.append(infoMovimiento['move']['url'])

    urlHabitat = f'{baseHabitat}{idPokemon}/'
    respuestaHabitat = requests.get(urlHabitat)
    habitat = "no encontrado"
    if respuestaHabitat.status_code == 200:
        datosHabitat = respuestaHabitat.json()
        if 'habitat' in datosHabitat:
            habitat = datosHabitat['habitat']['name']
    
    return {
        'Nombre': nombre,
        'Movimientos': movimientos,
        'Habitat': habitat 
    }
    
def obtenerMovimientos(urlMovimiento):
    respuesta = requests.get(urlMovimiento)
    if respuesta.status_code == 200:
        datosMovimiento = respuesta.json()
        nombreMovimiento = datosMovimiento["name"]
        if datosMovimiento["effect_entries"]:
            efecto = datosMovimiento["effect_entries"][0]["effect"]
            efectoCorto = datosMovimiento['effect_entries'][0]['short_effect']
        
        else:
            efecto = datosMovimiento['damage_class']['name']
            efectoCorto = 'N/A'
        return {
            'Nombre Movimiento': nombreMovimiento,
            'Efecto': efecto,
            'Efecto Corto': efectoCorto
        }
    else:
        return {'Nombre Movimiento': 'Desconocido', 'Efecto': 'Desconocido', 'Efecto Corto': 'Desconocido'}
    
    
    
for idPokemon in idsPokemons:
    infoPokemon = obtenerInfoPokemon(idPokemon)
    if infoPokemon.get('Movimientos'):
        movimientos = infoPokemon['Movimientos']
        
        for urlMovimiento in movimientos[:1]:
            detallesMovimiento = obtenerMovimientos(urlMovimiento)
            detallesMovimiento['Nombre Pokemon'] = infoPokemon['Nombre']
            detallesMovimiento['Habitat'] = infoPokemon['Habitat']
            listaMovimientos.append(detallesMovimiento)
            
dataframeMovimientos = pd.DataFrame(listaMovimientos)
dataframeMovimientos = dataframeMovimientos.rename(columns={
    
    'Nombre Movimiento': 'MOVIMIENTO',
    'Nombre Pokemon': 'POKEMON',
    'Efecto': 'EFECTOS',
    'Efecto Corto': 'EFECTOS CORTOS',
    'Habitat': 'HABITAD'
})

dataframeMovimientos = dataframeMovimientos[['POKEMON', 'MOVIMIENTO', 'EFECTOS', 'EFECTOS CORTOS', 'HABITAD']]
print(tabulate(dataframeMovimientos, headers='keys', tablefmt='pretty'))
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    

 
        
    






```
